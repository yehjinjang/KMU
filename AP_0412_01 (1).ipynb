### Enumerate

<font color = "blue">
Ex) enumerate()를 이용하여 ['a', 'b', 'c']라는 리스트를 {0:'a', 1:'b', 2:'c'}라는 딕셔너리로 바꾸시오.

````python
lst = ['a','b','c']
adict = {}  # 빈 딕셔너리를 만든다.
````

````python
for i,v in enumerate(lst):
    adict[i] = v  # 딕셔너리에 새로운 원소를 추가하려면 => 변수명[key] = value
````

````python
adict
````

````python
# dict comprehension 까지 사용하면 보다 간단히 표현할 수 있음.
{i: v for i,v in enumerate(lst)}
````

### List Comprehension

<font color = "blue">
Ex) 아래 딕셔너리 변수를 값 기준으로 정렬하시오.

````python
c = {'a': 10, 'c': 1, 'b': 22}
````

````python
# 아래 코드는 키값을 기준으로 정렬됨 (X)
sorted(c.items())  
````

````python
# 먼저, list comprehension을 이용하여 (key, value) -> (value, key)로 변경
[(v,k) for k, v in c.items()]
````

````python
# 다음으로, sorted 내장함수를 이용하여 값 기준으로 정렬한다.
sorted([(v,k) for k, v in c.items()])
````

````python
sorted([(v,k) for k, v in c.items()], reverse=True)
````

### 문제 16)
다음은 names라는 리스트에 들어있는 단어와 그것의 횟수를 출력하는 코드이다.
```
counts = dict()
names = ['빅데이터', '경영', '통계', '경영', '경영', '정보']
for name in names :
    counts[name] = counts.get(name, 0) + 1
print(counts)
```
위의 코드를 응용하여, 오바마 연설문에서 가장 많이 나타나는 단어와 그것의 횟수를 출력하는 프로그램을 작성하시오.

````python
with open('Obama_words.txt', 'r') as f:
    words = f.read()
    print(words)
    f.close
   
# words에서 단어를 구분
words = words.split()
words

# 위 코드를 이용하여 각 단어와 그것의 횟수를 계산
counts = dict()
for word in words :
    counts[word] = counts.get(word,0) +1
#    print(counts) # 이 라인을 막으면 내가 말한 메시지가 안나온다.

# 가장 자주 언급되는 단어를 찾고 그것의 횟수를 출력
y = 0
for w,c in counts.items():
    if y < c:
        y = c
        m = w
print(m, y)      
````

<font color="darkgreen"><p>
#### Revision #1: list comprehension과 max() 함수를 사용하여 가장 많이 나타나는 단어와 그것의 횟수를 출력

````python
[(v,k) for k,v in counts.items()]
````

````python
max([(v,k) for k,v in counts.items()])
````

#### Revision #2: list comprehension과 sorted() 함수를 사용하여 출현 빈도가 가장 높은 상위 10개 단어 리스트 출력

````python
sorted([(v,k) for k,v in counts.items()], reverse=True)
````

````python
sorted([(v,k) for k,v in counts.items()], reverse=True)[:10]
````
