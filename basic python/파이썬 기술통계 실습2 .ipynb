```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
Brand = pd.read_excel ("./SoftDrink.xlsx", header= 0, usecols="A")
Brand.columns = Brand.columns.str.replace(' ','_')
Brand
```

```python
Brand.Brand_Purchased.value_counts() #종류의 갯수들 세기 
```

```python
Brand_Frequency = pd.DataFrame(Brand.Brand_Purchased.value_counts()) 
Brand_Frequency.reset_index(inplace=True) #reset_index index를 새로만들어주는것
Brand_Frequency
```

```python
Brand_Frequency.columns =["Name","freq"]
Brand_Frequency
```

```python
Brand_Frequency["ratio"] = np. round(Brand_Frequency.freq/sum(Brand_Frequency.freq),2)
```

```python
Brand_Frequency["percentage"] = np.round(Brand_Frequency.freq/sum(Brand_Frequency.freq)*100,2).astype(str) +'%' 
#astype에 대한 결과를 str이라는 속성으로 바꿔줘라 문자로 바꾼 다음 뒤에 퍼센테이리즐 붙여야함 
```

```python
Brand_Frequency
```

```python
n = len(Brand_Frequency['freq'])
```

```python
ax = plt.subplot(111) #ax 그래프를 그리라는 함수 
plt.bar((range(n)),Brand_Frequency['freq'])
ax.set_xticks(range(n)); #x축 눈금 만들기 
ax.set_xticklabels(Brand_Frequency['Name'])
```
