```python
def split_sentences(text):    # 하나의 문서를 여러문장으로 나누기
    import re    
    all_sentences = []    
    lines = [line for line in text.strip().splitlines() if line.strip()]
    
    for line in lines:
        sentences = re.split("(?<=[.?!])\s+", line)
        all_sentences += sentences
    
    return all_sentences



def get_sentences(input_file_name):  #파일 불러오기
    
    sentences = []
    
    with open(input_file_name, "r", encoding="utf-8") as input_file:
        for sentence in input_file:
            sub_sentences = split_sentences(sentence)  
            sentences += sub_sentences
    
    return sentences



def get_parts(sentences):   #komoran 실행
    
    from konlpy.tag import Komoran
    
    komoran = Komoran()
    results = []
    
    for sentence in sentences:
        parts = komoran.pos(sentence)
        for part in parts:
            results.append(part)
    
    return results


def filtered_parts(parts): 
    
    N_POS = ["NNG", "NNP"]  
    V_POS = ["VV", "VA"]   
    X_POS = ["XR"]          
    M_POS = ["MAG", "MAJ"]   
    
    filtered = []
    
    for part in parts:
        if part[1] in N_POS:
            filtered.append(part[0])
        elif part[1] in V_POS:
           filtered.append(part[0]+"다") # 다가 필요한 경우
        elif part[1] in X_POS:
            filtered.append(part[0]+"하다") # 하다 가 필요한 경우
        elif part[1] in M_POS:
            filtered.append(part[0])
        else:
            continue                
    return filtered


def word_count(filtered):    #형태소 전체 빈도 확인 
    from collections import Counter    
    wordform_counter = Counter()

    for word in filtered:
        wordform_counter[word] += 1    
   
    for word, count in wordform_counter.most_common(20): 
        print("{}: {}".format(word, count))    
    
def main():
    input_file_name = "omicron.txt"
    sentences = get_sentences(input_file_name)     
    parts = get_parts(sentences)     
    filtered = filtered_parts(parts)     
    word_count(filtered)     

 
main()

```

- Komoran모듈을 이용한 형태소 분석에서는 오미크론에서 오미와 크론의 '크'를 동사(VV)로 인식하였고, 론은 명사파생 접미사(XSN)로 인식 되어 오미크론이 총 26개로 가장 많은 명사 형태소로 추출 되었다.

- twitter(Okt)모듈을 이용한 형태소 분석에서는 '오미크론'을 하나의 명사로 인식하여 총 26라는 것을 다시 확인 할 수 있었다. 2번째로 많은 빈도를 차지한 '확진자'라는 명사는 앞선 komoran의 분석과는 다르게 '확'noun, '진자'noun인 형태소로 인식 되었다.
