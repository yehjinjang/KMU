```python
import csv 
f = open('age.csv')
data = csv.reader(f)
result =[]

name = input('인구 구조가 알고싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))



plt.plot(result)
plt.title(name+' 지역의 인구 구조')
plt.show()
```

```python
df = pd.read_csv('age.csv', index_col = 0, thousands = ',')
```

```python
#연령별 인구수 비율 
df2 =df.apply(lambda x: x/df['2019년02월_계_총인구수'])
del df2['2019년02월_계_총인구수'],df2['2019년02월_계_연령구간인구수']
df2.head()
```

```python
name = input('원하는 지역명')
a= df2.index.str.contains(name) #인덱스문자열에 name에 입력한 값을 포함한 행 찾기
df3 = df2[a]
df3
#원하는 지역의 인구 비율 
```

```python
df3.T.plot()
plt.show()
```

```python
#선택한 지역과 비슷한 인구구조를 가진 지역 시각화 
x = df2.sub(df3.iloc[0],axis=1)
y= np.power(x,2)
z = y.sum(axis=1)

i = z.sort_values().index[:5] #차이가 적은 지역 5곳 추출 

df2.loc[i].T.plot()
plt.show()
```
